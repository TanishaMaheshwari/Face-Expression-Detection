```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, BatchNormalization, Flatten, Dense
from tensorflow.keras.applications import MobileNet

# Load the base MobileNet model without the top layers
base_model = MobileNet(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

# Define the complete model architecture
model = Sequential([
    base_model,
    Dropout(0.25),
    BatchNormalization(),
    Flatten(),
    Dense(1024, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(10, activation='softmax')  # Adjust this if your model has a different number of classes
])

# Compile the model (necessary before loading weights)
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Load the weights into the model
try:
    model.load_weights('/Users/tanishamaheshwari/Downloads/8thsem_project.h5')
    print("Weights loaded successfully.")
except ValueError as e:
    print(f"Error loading weights: {e}")

# Check the model summary to verify the architecture
model.summary()
```
